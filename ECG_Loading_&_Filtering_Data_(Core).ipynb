```python
# open the excel file RailsToTrails National Count Data_week.xlsx
rails = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/RailsToTrails_National Count Data_week.xlsx')
```

```python
# check the first 10 rows of the dataframe
rails.head()
```

```python
# select only data after july 31 2020
july31_filter = rails["Week of"] > '2020-07-31'
july31_filter.head()
```

```python
rails.loc[july31_filter, :].head()
```

```python
# select data where the 'Change 2019-2020' is greater than 1
change_filter = rails['Change 2019-2020'] > 1
change_filter.head()
```

```python
rails.loc[change_filter, :].head()
```

```python
# combine both columns (how only the header)
rails_filter = rails.loc[july31_filter & change_filter, :]
rails_filter
```
