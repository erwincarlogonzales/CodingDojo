```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# import file
filename = ('/content/drive/MyDrive/Colab Notebooks/titanic.csv')
df = pd.read_csv(filename)
```

```python
# check for null values
# we can see that some values in Age and Cabin are NA
df.isna()
```

```python
# what % of people survived (hint: find the mean of the survival column)
survivors = df['Survived'].sum()
total_passengers = df['PassengerId'].count()

percent_survivors = survivors / total_passengers
percent_survivors
```

```python
# what % of people survived (hint: find the mean of the survival column)
survivors_mean = df['Survived'].mean()
survivors_mean
```

```python
# how many of each sex survived?
df['Sex'].value_counts()
```

```python
# filter all male passengers
male_filter = df['Sex'] == 'male'
df.loc[male_filter].head()
```

```python
# filter all survived passengers
survived_filter = df['Survived'] == 1
df.loc[survived_filter]
```

```python
# filter both male and survived passengers
male_survivors = df.loc[male_filter & survived_filter]
male_survivors.count()
```

```python
# filter all male passengers
female_filter = df['Sex'] == 'female'
df.loc[female_filter].head()
```

```python
# filter both female and survived passengers
female_survivors = df.loc[female_filter & survived_filter]
female_survivors.count()
```

```python
# male survivors are 109 while female survivors were 233
male_survivors.count(), female_survivors.count()
```

```python
# overall, what is the average age of those who died?
# filter all those who did not survive
perished = df['Survived'] == 0
df[perished].count()

# get the mean of their age
perished_age = df.loc[perished].mean()
perished_age
```

```python
# filter those who perished
perished_filter = df.loc[perished]
perished_filter
```

```python
# create an Age filter just for the fuck of it
age_filter = df['Age']
age_filter

# create a filter for those who perished that includes a god damn age filter
perished_age = df.loc[perished & age_filter]
perished_age.head()
```

```python
# average age of those who perished or did not survive
perished_age.groupby(['Survived']).mean('Age')
```

```python
# overall, what is the average age of those who survived? there are null values in Age
survived_age = df.loc[survived_filter & age_filter]
survived_age.groupby(['Survived']).mean('Age')
```

```python
# what is the average age of those who did survive by sex?

# ave age of male survivors
male_survived_age = df.loc[male_filter & survived_filter]
male_survived_age.groupby(['Sex']).mean('Age')
```

```python
# ave age of female survivors
female_survived_age = df.loc[female_filter & survived_filter]
female_survived_age.groupby(['Sex']).mean('Age')
```

```python
# what is the average of of those who died by sex?

# ave age of males who perished
male_perished_age = df.loc[male_filter & perished]
male_perished_age.groupby(['Sex']).mean('Age')
```

```python
# ave age of females who perished
female_perished_age = df.loc[female_filter & perished]
female_perished_age.groupby(['Sex']).mean('Age')
```
