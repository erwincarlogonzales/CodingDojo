```python
# import libraries
import pandas as pd
```

```python
# read the dataset using url
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data', header = None)
df.head()
```

```python
# check the dataset
df.info()
```

```python
# rename the columns
# rename the following
# column 0 to 'Flow_ID'
# column 1 to 'Date'
# column 2 to 'Time'
# column 3 to 'Count'

df = df.rename(columns = {0: 'Flow_ID',
                          1: 'Date',
                          2: 'Time',
                          3: 'Count'})
df.head()
```

```python
# select all data with date July 24 2005 with flow id = 7
date_filter = df['Date'] == '07/24/05'
df[date_filter]
```

```python
# select all data with date July 24 2005 with flow id = 7
flow_filter = df['Flow_ID'] == 7
df[flow_filter]
```

```python
# apply both flow_ID and Date filters
flow_date = df.loc[date_filter & flow_filter]
flow_date.head()
```

```python
# from data selected above sort the result in Count in descending order and output the top 10 rows. assign the output to the variabe top10
# df.sort_values(by='col1', ascending=False)

top10 = flow_date.sort_values(by = 'Count', ascending = False).head(10)
top10
```

```python
# For the 10 rows outputted above, use Pandas Apply function to subtract lowest value of the 10 from all of them and then output the average value of the resulting counts

# subtract the lowest value (1) form the Count of all values in top10
# return the Count mean()

count_min = top10['Count'].min()

top10_mean = top10['Count'].apply(lambda x: x - count_min).mean()
top10_mean
```

```python
# Grouping
# Select data in the month of August 2005 having flow id=7
# aug_filter = df['Date'].between('08/01/05', '08/31/05')
# df[aug_filter]

aug_start = df['Date'] >= '08/01/05'
aug_end = df['Date'] <= '08/31/05'
aug_filter = aug_start & aug_end
aug_filter
```

```python
# filter aug_filter by flow id = 7
flow_aug = df[aug_filter & flow_filter]
flow_aug
```

```python
# Group the data based on date and get the max count per date
flow_aug.groupby(['Date']).max('Count')
```
