```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from pandas.core.arrays import categorical
```

```python
# import file
filename = '/content/drive/MyDrive/Colab Notebooks/insurance.csv'
df = pd.read_csv(filename)
df.head()
```

How well can the charge be predicted based on the age, sex, bmi, number of children, smoking habit and region of the patient?

```python
# Define features (X) and target (y)
y = df['charges']
X = df.drop(columns = ['charges'])
```

```python
# Train test split the data to prepare for machine learning
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

Identify each feature as numerical, ordinal, or nominal. (Please provide this answer in a text cell in your Colab notebook)

numerical = age, bmi

ordinal

nominal = sex, children, region

```python
# Ordinal encode any ordinal features
# there are no ordinal features for this dataset
```

```python
# One Hot Encode any nominal features
# create selectors
cat_selector = make_column_selector(dtype_include = 'object')
num_selector = make_column_selector(dtype_include = 'number')

# select categorical and numerical columns
cat_selector(X_train)
num_selector(X_train)
```

```python
# create subset for categorical cols
train_cat_data = X_train[cat_selector(X_train)]
test_cat_data = X_test[cat_selector(X_test)]
train_cat_data
```

```python
# create subset for numerical cols
train_num_data = X_train[num_selector(X_train)]
test_num_data = X_test[num_selector(X_test)]
train_num_data
```

```python
# Ordinal encode any categorical features
ohe = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
ohe.fit(train_cat_data)
```

```python
# Scale any numeric features
scaler = StandardScaler()
scaler.fit(train_num_data)
```

```python
# transform categorical train and test data
train_ohe = ohe.transform(train_cat_data)
test_ohe = ohe.transform(test_cat_data)
train_ohe
```

```python
# transform numerical train and test data
train_num = scaler.transform(train_num_data)
test_num = scaler.transform(test_num_data)
train_num
```

```python
# compose header names for numerical data
# create numerical data col names using list
num_colNames = ['age', 'bmi', 'children']

# create dataframe from numerical array
train_num_df = pd.DataFrame(train_num, columns = num_colNames)
test_num_df = pd.DataFrame(test_num, columns = num_colNames)
train_num_df
```

```python
# Return all column names to the combined dataframe
# get column names of categorical training set
ohe_column_names = ohe.get_feature_names(train_cat_data.columns)
train_ohe_df = pd.DataFrame(train_ohe, columns = ohe_column_names)
test_ohe_df = pd.DataFrame(test_ohe, columns = ohe_column_names)
train_ohe_df
```

```python
# Concatenate all features back into one dataframe
# combine transformed train numerical and categorical data
trained_final = pd.concat([train_num_df, train_ohe_df], axis=1)
trained_final
```

```python
# combine transformed test numerical and categorical data
test_final = pd.concat([test_num_df, test_ohe_df], axis=1)
test_final
```

For this task, you will need to:

Instantiate a linear regression model

Fit the model on the training data

Evaluate the model performance on both the training set and the test set using 
R^2 score

```python
# Instantiate a linear regression model
reg = LinearRegression()
```

```python
# train the model on training data (it attempts to learn the relationship between X and y)
# Fit the model on the training data
reg.fit(trained_final, y_train)
```

```python
# Evaluate the model performance on both the training set and the test set using R^2 score
trained_score = reg.score(trained_final, y_train)
print(trained_score)
```

```python
tested_score = reg.score(test_final, y_test)
print(tested_score)
```

```python
# obtain predictions
# results of the train set
trained_pred = reg.predict(trained_final)

# results of the test set
test_pred = reg.predict(test_final)

# display values of the train and test predictions
print(trained_pred, test_pred)
```
