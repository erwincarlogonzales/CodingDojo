```python
# Import the data as a Pandas DataFrame.
import pandas as pd
```

```python
# load the file Name the DataFrame bike_df
filename = '/content/drive/MyDrive/Colab Notebooks/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(filename)
```

```python
# check the dataframe
bike_df.head()
```

```python
# delete unnecessary columns
bike_df.drop('Unnamed: 7', inplace = True, axis = 1)
bike_df.drop('Unnamed: 8', inplace = True, axis = 1)
bike_df.drop('Unnamed: 9', inplace = True, axis = 1)
```

```python
# check the dataframe
bike_df.head()
```

```python
# rename unnamed: 6 to percent change
bike_df = bike_df.rename(columns = {'Unnamed: 6': 'percent_change'})
bike_df.head()
```

```python
# count the missing values in the dataset and which columns are they in
bike_df.isnull().sum(axis = 0)
```

```python
# fill the missing values in 2021 Counts with 0
bike_df['2021 Counts'] = bike_df['2021 Counts'].fillna(0)
bike_df.head()
```

```python
# rename 2021 Counts to counts_2019 >
# there is a (space before the 2019 counts (31 counters) make sure to select it)
# this can be seen when bike_df.columns
bike_df.columns
```

```python
# rename columns 2019, 2020, 2021
bike_df = bike_df.rename(columns = {' 2019 counts (31 counters)': 'counts_2019',
                                    ' 2020 counts (31 counters)': 'counts_2020',
                                    '2021 Counts': 'counts_2021'})
bike_df.head()
```
