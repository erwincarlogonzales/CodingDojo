```python
# import libraries
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import file
filename = '/content/drive/MyDrive/Colab Notebooks/Cost of 1GB of Data.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
# set variables
rank = df['Rank'].values
country = df['Country'].values
cost = df['Avg Price of 1GB (USD)'].values
```

```python
# bar chart
plt.figure(figsize = (25, 15))
plt.xticks(rotation = 90);
plt.bar(country, cost);
```

```python
# OO bar chart
fig, axes = plt.subplots(figsize = (25,15),
                         nrows = 1,
                         ncols = 1,)
axes.bar(country, cost)
plt.xticks(rotation = 90);
# axes.legend(loc = 'upper left');
```
